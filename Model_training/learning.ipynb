```python
import matplotlib.pyplot as plt
```

```python
import librosa
import IPython.display as ipd
import librosa.display
```

```python
stutter_clip1 = "stutter_audio/HeStutters/1/HeStutters_1_3.wav"
stutter_clip2 = "stutter_audio/HeStutters/1/HeStutters_1_4.wav"
stutter_clip3 = "stutter_audio/HeStutters/1/HeStutters_1_5.wav"

clean_clips1 = "clean_audio/HeStutters/1/HeStutters_1_4.wav"
clean_clips2 = "clean_audio/HeStutters/1/HeStutters_1_19.wav"
clean_clips3 = "clean_audio/HeStutters/1/HeStutters_1_52.wav"
```

```python
stutter_files = [stutter_clip1, stutter_clip2, stutter_clip3]

clean_files = [clean_clips1,clean_clips2,clean_clips3]
```

```python
for file in stutter_files:
    y, sr = librosa.load(file, sr=None)

    plt.figure(figsize=(14, 5))
    librosa.display.waveshow(y, sr=sr)
    plt.title(file)
    plt.show()

    display(ipd.Audio(file))
```

```python
for file in clean_files:
    y, sr = librosa.load(file, sr=None)

    plt.figure(figsize=(14, 5))
    librosa.display.waveshow(y, sr=sr)
    plt.title(file)
    plt.show()

    display(ipd.Audio(file))
```
